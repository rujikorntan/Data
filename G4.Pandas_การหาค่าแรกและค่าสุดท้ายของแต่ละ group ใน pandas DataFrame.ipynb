```python
pvt = df2.pivot_table(index='school', columns='acad_year', values='cost')
```

```python
pvt
```

$$
\begin{aligned}
a(1+r)^n &= b \\
(1+r)^n &= \frac{b}{a} \\
(1+r) &= \big(\frac{b}{a}\big)^\frac{1}{n} \\
r &= \big(\frac{b}{a}\big)^\frac{1}{n} - 1
\end{aligned}
$$

```python
pvt.columns
```

```python
period = pvt.columns[1] - pvt.columns[0]          #หาค่า n
```

```python
period
```

```python
pvt.loc[:, 2017]/pvt.loc[:, 1985]
```

```python
(pvt.loc[:, 2017]/pvt.loc[:, 1985])**(1/period) - 1 #ค่าเทอมเพิ่มขึ้น กี่ % ต่อปี
```

```python
pvt['cagr'] = (pvt.loc[:, 2017]/pvt.loc[:, 1985])**(1/period) - 1 
```

```python
pvt
```

```python
pvt['cagr'].sort_values().plot.barh();
```

### วิธีที่2

```python
def cagr (a, b, peroid):
    '''
    Cumulative Annual Growth Rae(CAGR)
    a = value in the first year
    b = value int the last year
    peroid = number of years between the first and the last
    '''
    return (b/a)**(1/period)-1
```

```python
pvt.apply(lambda v : cagr(v[1985], v[2017], period), axis=1)
```

```python
pvt['cagr2'] = pvt.apply(lambda v : cagr(v[1985], v[2017], period), axis=1)
```

```python
import pandas as pd
```

```python
df = pd.read_csv('https://github.com/prasertcbs/tutorial/raw/master/msleep.csv', nrows=10)
df
```

```python
def hilight_font(e):
    color = 'red' if e > 10 else 'black'
    return f'color : {color}'
```

```python
hilight_font(11)
```

```python
df.style.applymap(hilight_font, subset = ['sleep_total'])
```

```python
def hilight_bg(e):
    return 'background-color: yellow' if e > 10 else ''
```

```python
df.style.applymap(hilight_bg, subset=['sleep_total'])
```

```python
df.style.applymap(lambda e: f'background-color:deepskyblue' if e > 10 else '', subset=['awake'])
```

```python
def upfont (e ,v):
    if e > v :
        style= 'font-size: 150%; background-color:deepskyblue; color:red; border: 3px solid yellow'
    else:
        style=''
    return style
```

```python
df.style.applymap(upfont, v=10, subset=['sleep_total'])
```

```python
df.style.applymap(lambda e: 'background-color:deepskyblue; color:red; font-size:130%; border: 3px solid blue' if e>10 else '',
                 subset=['sleep_total']) 
```

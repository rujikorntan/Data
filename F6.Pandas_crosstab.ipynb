```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('https://github.com/prasertcbs/tutorial/raw/master/mpg.csv')
df
```

```python
pd.crosstab(df.drv, df['class'])
```

```python
pd.crosstab(df.drv, df['class'], margins=True)
```

```python
pd.crosstab(df['drv'], df['class'], margins=True, normalize=True)
```

```python
pd.crosstab(df['drv'], df['class'], margins=True, normalize=True) *100
```

```python
(pd.crosstab(df['drv'], df['class'], margins=True, normalize=True)*100).round(1)
```

```python
df['trans_am'] = df['trans'].str[0]
df
```

```python
pd.crosstab(df.trans_am, df.manufacturer)
```

```python
pd.crosstab([df['class'], df.trans_am], df.manufacturer, margins=True)
```

```python
df.agg(['mean', 'max', 'min'])
```

```python
pd.crosstab([df['class'], df.trans_am], df.manufacturer, margins=True, values=df.cty, aggfunc=np.average) #ใส่ค่าเฉลี่ยของ cty
```

```python
pd.crosstab([df['class'], df.trans_am], df.manufacturer, margins=True, values=df.cty, aggfunc=np.average).sort_index().loc[['compact', 'minivan']]
```

```python
pd.crosstab([df['class'], df.trans_am], df.manufacturer, margins=True, values=df.cty, aggfunc=np.max).loc[['compact', 'minivan']]
```

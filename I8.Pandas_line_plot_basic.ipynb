```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter
%matplotlib inline
%config InlineBlackend.figure_format = 'retina'
```

```python
pd.set_option('max_rows', 10)
```

```python
df = pd.read_csv('https://github.com/prasertcbs/basic-dataset/raw/master/OPEC_monthly_oilprice.csv',
                 parse_dates=True, index_col='Date')
df
```

```python
df.plot(color='red')
```

```python
ax = df.plot(kind='line', title='OPEC oil price US$/barrel', color='maroon')
ax.set_ylabel('US$/barrel')
```

```python
ax = df.plot(kind='line', title='OPEC oid price US$/barrel', color='red')
ax.axhline(df.Oil.mean(), linestyle='--', linewidth=1, color='blue')
ax.set_xlabel('')
ax.set_ylabel('US$/barrel')
```

```python
ax = df.loc['2017-01-31':,:].plot(kind='line', title='OPEC oil priece US$/barrel', color='red')
ax.axhline(df.loc['2017-01-31':,'Oil'].mean(), linestyle='--', color='black', linewidth=1 )
ax.set_xlabel('')
ax.set_ylabel('US$/barrel')
```

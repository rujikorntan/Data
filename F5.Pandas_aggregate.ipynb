```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/prasertcbs/tutorial/master/mpg.csv')
df
```

```python
df.describe()
```

```python
df.agg(['mean'])
```

```python
df.agg(['min', 'max', 'mean'])
```

```python
df.cty.mean()
```

```python
df[['cty', 'hwy']].agg(['min', 'max', 'mean'])
```

```python
df.groupby('year').mean()
```

```python
df.groupby(['class', 'year']).mean()
```

```python
df.groupby(['class', 'year'])[['cty', 'hwy']].mean()
```

```python
df.groupby(['class', 'year'])[['cty', 'hwy']].agg(['max', 'min'])
```

```python
df.groupby('class')['cty'].mean()  # df.groupby('class').cty.mean()
```

```python
df.groupby('class')['cty'].mean().sort_values().plot.barh()
```

```python
avg = df.groupby('class')['cty'].mean().mean()
avg
```

```python
df.groupby('class')['cty'].mean().sort_values().plot.barh(color='deepskyblue')
plt.axvline(x=avg, linestyle='--', color='red')                     # เพิ่ม vertical line
```

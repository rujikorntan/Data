```python
import pandas as pd
```

```python
dfs = pd.read_html('https://en.wikipedia.org/wiki/List_of_tallest_buildings', match ='Burj Khalifa', header=0)
```

```python
len(dfs)
```

```python
dfs[1]
```

```python
df = dfs[1]
```

```python
df
```

```python
df['Building'] = df.Building.str.replace('†','')
```

```python
df.rename(columns={'Unnamed: 7':'Year'}, inplace=True)
```

```python
df
```

```python
df.info()
```

```python
df['Height'] = df.Height.str.replace('(\[.*\])|[^\d\.]','').astype(float)
```

```python
df
```

```python
df['Floors'] = df.Floors.str.replace('\[.*\]|[^\d]','').astype(int)
df
```

```python
df.info()
```

```python
df['Rank_floor'] = df.Floors.rank(method='min', ascending=False).astype(int)
```

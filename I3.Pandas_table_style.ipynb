```python
import pandas as pd
```

```python
df = pd.read_csv('https://github.com/prasertcbs/tutorial/raw/master/msleep.csv', nrows=10)
df
```

```python
df.style.highlight_null('pink')
```

```python
df.style.highlight_max(color='blue')
```

```python
df.style.highlight_max(color='green').highlight_min(color='red')
```

```python
df.style.highlight_max(color='green').highlight_min(color='red').highlight_null('gray')
```

```python
df.style.bar()
```

```python
df.style.bar(color='deepskyblue', subset=['sleep_total', 'awake'])
```

```python
df.sort_values('sleep_total', ascending=False).style.bar(color='deepskyblue', subset=['sleep_total', 'awake'])
```

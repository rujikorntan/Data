# Inner join

```python
df = pd.merge(maths, sci, on='Country', suffixes=['_math', '_science'])    #เชื่อมด้วย Country
```

```python
df
```

```python
df = pd.merge(maths[['Country', 'Maths']], 
             sci[['Country', 'Science']],
             on='Country')
```

```python
df
```

```python
df = pd.merge(df, reading[['Country', 'Reading']], on='Country')
```

```python
df
```

## ใช้ merge 3 cell

```python
df =  pd.merge(pd.merge(maths[['Country', 'Maths']], 
                 sci[['Country', 'Science']],
                 on='Country') , 
               reading[['Country', 'Reading']] )
```

```python
df
```

```python
df['Total'] = df.Maths + df.Science + df.Reading
```

```python
df
```

```python
df['Rank'] = df.Total.rank(method='dense', ascending=False).astype(int)
```

```python
df.sort_values('Rank')
```

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('https://github.com/prasertcbs/tutorial/raw/master/mpg.csv')
df.sample(10)
```

```python
df.cty.plot.box();
```

```python
df[['cty', 'hwy']].boxplot(grid= False);
```

```python
df.boxplot(column = ['cty', 'hwy'])
```

```python
df.boxplot(grid = False, column = ['cty', 'hwy'], by ='drv');
```

```python
df.boxplot(grid = False, column = ['cty'], by = ['class', 'drv']);
```

```python
df.boxplot(grid = False, column = ['cty'], by=['class', 'drv'], rot = 45);
```

```python
df.boxplot(grid=False, column=['cty'], by=['class', 'drv'], vert=False);
```

```python
df.boxplot(grid=False, column=['cty', 'hwy'], by=['class', 'drv'], vert=False, figsize=(15,5));
```

```python
import pandas as pd
```

```python
west = pd.read_csv('https://github.com/prasertcbs/basic-dataset/raw/master/province_in_region/W.csv')
west.head()
```

```python
west.to_excel('datasheet/province_west.xlsx')
```

```python
west.to_excel('datasheet/province_west.xlsx', index=False)
```

```python
east = pd.read_csv('https://github.com/prasertcbs/basic-dataset/raw/master/province_in_region/E.csv')
east.head()
```

```python
south = pd.read_csv('https://github.com/prasertcbs/basic-dataset/raw/master/province_in_region/S.csv')
south.head()
```

```python
north = pd.read_csv('https://github.com/prasertcbs/basic-dataset/raw/master/province_in_region/N.csv')
north.head()
```

```python
df = pd.concat([north, south, west, east], keys=['ภาคเหนือ', 'ภาคใต้', 'ภาคตะวันออก', 'ภาคตะวันตก'])
df
```

##  To Excel

```python
df.to_excel('datasheet/province_merge.xlsx')
```

```python
df.to_excel('datasheet/province_merge.xlsx', merge_cells=False)
```

```python
df.to_excel('datasheet/province_merge.xlsx', merge_cells=False, index_label=['region', 'id'], sheet_name='province')
```

```python
excel_file = pd.ExcelWriter('datasheet/province_merge_sheet.xlsx')
df.to_excel(excel_file, sheet_name='ALL', index_label=['region','id'], merge_cells=False)
excel_file.close()
```

## Save each dataframe to separater excel sheet

```python
excel_file = pd.ExcelWriter('datasheet/province_each_sheet.xlsx')
north.to_excel(excel_file, sheet_name='NORTH', merge_cells=False, index_label=['region', 'id'])
south.to_excel(excel_file, sheet_name='SOUTH', merge_cells=False, index_label=['region', 'id'])
west.to_excel(excel_file, sheet_name='WEST', merge_cells=False, index_label=['region', 'id'])
east.to_excel(excel_file, sheet_name='EAST', merge_cells=False, index_label=['region', 'id'])
excel_file.close()
```

## use list or for

```python
sheets=['North', 'South', 'West', 'East', 'All']
excel_file = pd.ExcelWriter('datasheet/province_each_sheet_list.xlsx')
[c.to_excel(excel_file, sheet_name=sheets[i], index_label=['region', 'id'], merge_cells=False) 
 for i,c in enumerate([north, south, west, east, df])]
excel_file.close()
```

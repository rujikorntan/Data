# Read multiple CSV 

```python
stocks = ['bbl', 'ktb', 'scb']
dfs = []
location_prefix = 'https://github.com/prasertcbs/tutorial/raw/master/eod/'
dfs = [pd.read_csv(f'{location_prefix}{s}.csv', parse_dates=True, index_col =0) for s in stocks]    # f'' string
```

```python
type(dfs)
```

```python
dfs[0]
```

```python
dfs[1]
```

```python
df = pd.concat(dfs)      #รวมไฟล์
```

```python
df
```

```python
df = pd.concat(dfs, keys = stocks)
df
```

```python
df = pd.concat(dfs, keys = stocks, names=['Symbols','Date'])
df
```

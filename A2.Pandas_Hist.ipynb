```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('https://github.com/prasertcbs/tutorial/raw/master/mpg.csv')
df.head()
```

```python
df.hist()
```

```python
df.cty.hist();  #ใส่ ; เพื่อปิดคำอธิบาย
```

```python
df.cty.hist(bins = 20);    #จำนวนแท่ง
```

```python
df[['cty', 'hwy']].hist(bins = 5, grid = False, color = 'green', sharex = True, sharey = True);
```

```python
df[['cty', 'hwy']].plot.hist();
```

```python
df[['cty', 'hwy']].plot.hist(alpha = 0.3);   #ดูส่วนที่ทับ/ซ้อนกัน
```

```python
df.cty.plot.density();    #df.cty.plot.kde(); 
```

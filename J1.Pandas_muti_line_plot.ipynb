```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter # for custom number format on x and y axis
%matplotlib inline
%config InlineBackend.figure_format = 'retina'
```

```python
pd.set_option('max_rows', 10)
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/prasertcbs/basic-dataset/master/oil_gold_setindex.csv',
                 index_col='Date', parse_dates=True)
df
```

```python
df.plot();
```

```python
df[['Gold', 'SET']].plot(kind='line', color=['deepskyblue', 'maroon']);
```

```python
df['Oil'].plot(kind='line', color='blue', title='OPEC oil price US$/barrel');
```

##  subplot

```python
df[['Oil', 'SET', 'Gold']].plot(subplots=True, color=['blue', 'red', 'green'], figsize=(14,4), layout=(1,3));
```

```python
fig, ax = plt.subplots(1, 3, sharex=False, sharey=False, figsize=(16,4))
cols = ['Gold', 'Oil', 'SET']
title = ['Gold US$/troy ounce', 'OPEC oil price US$/barrel', 'SET index']
color = ['red', 'green', 'blue']
plt.tight_layout()
for i in range(len(cols)):
    ax0 = df.loc[:,cols[i]].plot(color=color[i], title=title[i], ax=ax[i], kind='line')
    ax0.axhline(df[cols[i]].mean(), color='red', linestyle='--', linewidth=1)
    ax0.set_xlabel('');
    ax0.set_ylabel('');    
    ax0.yaxis.set_major_formatter(FuncFormatter(lambda v, p: "{:,.0f}".format(v)))
```

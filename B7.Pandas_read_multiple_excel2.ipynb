```python
import pandas as pd
%matplotlib inline
```

# Read single sheet

```python
columns = ['Brand', '1300cc', '1600cc', '1800cc', '2000cc', 'gt2011cc', 'elec', 'total']
df = pd.read_excel('datasheet/Car1_whole.xls', skiprows=7, names=columns, skipfooter=3, 
                   header=None, sheet_name=1, usecols='B:I')    #sheet_name=1 -> อ่านแถบที่1
df
```

# Read multiple sheet

```python
columns = ['Brand', '1300cc', '1600cc', '1800cc', '2000cc', 'gt2011cc', 'elec', 'total']
dfs = [pd.read_excel('datasheet/Car1_whole.xls', skiprows=7, skipfooter=3,
                     header=None, names=columns, usecols='B:I', sheet_name=r) for r in range(1,13)]
```

```python
dfs
```

```python
dfs[0]
```

```python
[n.set_index('Brand', inplace=True) for n in dfs]
```

```python
df = pd.concat(dfs, keys=range(1,13), names=['month', 'brand'])
df
```

```python
df.reset_index(inplace=True)
df.head()
```

# Save to excel 

```python
df.to_csv('datasheet/Car2.csv')
```

```python
sheets=['Jan', 'Feb', 'Mar', 'Apr', 'May', 'June', 'July', 'Aug', 'Sept', 'Oct', 'Nov', 'Dec']
excel_file = pd.ExcelWriter('datasheet/msheet2559.xlsx')
[c.to_excel(excel_file, sheet_name=sheets[i]) for i,c in enumerate(dfs)]
excel_file.close()
```

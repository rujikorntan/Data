```python
import pandas as pd
```

```python
df = pd.read_csv('https://github.com/prasertcbs/tutorial/raw/master/wide.csv')      # wide foemat
df
```

```python
pd.melt(df, id_vars='branch')          #id_vars = columnที่ไม่ถูกการ melt              
#long foemat
```

```python
df2 = pd.melt(df, id_vars='branch', var_name='menu', value_name='unit')
df2
```

```python
df2.groupby(['branch', 'menu']).describe()
```

```python
df.groupby('branch').describe()
```

## New Read

```python
columns = ['seq', 'Brand', '1300cc', '1600cc', '1800cc', '2000cc', 'gt2011cc', 'elec', 'total']
df = pd.read_excel('datasheet/Car1_whole.xls', sheet_name='Total',  skiprows = 7, skipfooter=3, header=None, names=columns)
df
```

```python
df.drop('seq', axis=1, inplace=True)  #ลบ column ที่ 1
df
```

```python
df.set_index('Brand', inplace=True)
df
```

```python
df = df[df.total!=0] # data อันใหม่เก็บ total != 0
df
```

```python
df.nlargest(10, 'total')
```

```python
df.nlargest(10, 'total')['total'].plot(kind='barh');
```

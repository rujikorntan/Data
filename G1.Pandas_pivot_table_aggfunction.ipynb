```python
import pandas as pd
import numpy as np
%matplotlib inline
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/prasertcbs/tutorial/master/long2.csv')
df
```

```python
df.pivot_table(index='branch', columns='menu', values='units', aggfunc='sum', margins=True, margins_name='Total')
```

```python
df.pivot_table(index='branch', columns='menu', values='units', aggfunc=[np.sum, np.mean],
              margins=True, margins_name='Total').round(2)
```

```python
df.pivot_table(index=['branch','period'], columns='menu', values='units',
               margins=True, margins_name='Total' , aggfunc='sum').round(2)
```

```python
df.pivot_table(index='branch', columns=['period', 'menu'], values='units',
              aggfunc='sum', margins=True, margins_name='Total')
```

```python
df.pivot_table(index='branch', columns=['menu', 'period'], values='units',
               margins=True, margins_name='Total').round(2).plot.barh()
```

```python
df.pivot_table(index='branch', columns='menu', values='units', aggfunc='sum').round(2).plot.barh()
```

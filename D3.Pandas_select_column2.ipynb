```python
import pandas as pd
```

```python
df = pd.read_csv('https://github.com/prasertcbs/tutorial/raw/master/score.csv', nrows=5)
df
```

```python
df['math_total'] = df.math1 + df.math2 + df.math3
df
```

```python
df.columns.str.match(r'^math\d{1,}')           # ^math=ขึ้นต้อนด้วย math , \d{1,} = ตัวเลข 1 หน่วย ขึ้นไป
```

```python
df.columns[df.columns.str.match(r'^math\d{1,}')]
```

```python
df[df.columns[df.columns.str.match(r'^math\d{1,}')]]
```

```python
df['math_total'] = df[df.columns[df.columns.str.match(r'^math\d{1,}')]].sum(axis=1)
df['eng_total'] = df[df.columns[df.columns.str.match(r'^eng\d{1,}')]].sum(axis=1)
df['sci_total'] = df[df.columns[df.columns.str.match(r'^sci\d{1,}')]].sum(axis=1)
df
```

```python
df['subject_total'] = df[df.columns[df.columns.str.match(r'\w+[123]$')]].sum(axis=1)     #axis=1 = ทำตาม column
df
```

```python
del df['subject_total']            #ลบ column
df
```

```python
df['subject_total'] = df[df.columns[df.columns.str.match(r'.*total$')]].sum(axis=1)  # .* = ตัวอะไรก็ได้
df
```

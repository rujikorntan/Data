```python
import pandas as pd
%matplotlib inline
```

```python
df = pd.read_table('https://github.com/prasertcbs/tutorial/raw/master/currency.tsv', index_col='Year')
df
```

```python
df.TV.str.replace('\D','').astype(int)      # \D = ที่ไม่ใช่ตัวเลข
```

```python
import re
def remove_nondigit(v):
    return int(re.sub(r'\D', '', v))
```

```python
remove_nondigit('asd1a23')
```

```python
df.applymap(remove_nondigit)
```

```python
df.applymap(lambda v: re.sub(r'\D', '', v))
```

```python
df2= df.applymap(remove_nondigit)
```

```python
df2
```

```python
df2.plot()
```

```python
import pandas as pd
```

```python
dfs = pd.read_html('https://en.wikipedia.org/wiki/Programme_for_International_Student_Assessment',
                 match='Singapore')
```

```python
maths= dfs[1]
sci = dfs[2]
reading = dfs[3]
```

```python
maths.rename(columns={0:'Rank', 1:'Country', 2:'Maths'}, inplace=True)
sci.rename(columns={0:'Rank', 1:'Country', 2:'Science'}, inplace=True)
reading.rename(columns={0:'Rank', 1:'Country', 2:'Reading'}, inplace=True)
```

```python
maths
```

```python
sci
```

```python
reading
```

```python
df = pd.merge(maths[['Country', 'Maths']], pd.merge(sci[['Country', 'Science']], reading[['Country', 'Reading']], on='Country'), on='Country')
```

```python
df
```

```python
df['Total'] = df.Maths + df.Science + df.Reading
```

```python
df
```

```python
df['Rank'] = df.Total.rank(method='dense', ascending=False).astype(int)
```

```python
df.sort_values('Rank', ascending=True, inplace=True)
```

```python
df
```

```python
pisa = df
```

## multicolumn calculate

```python
df_z = (df[['cty', 'hwy']] - df[['cty', 'hwy']].mean())/df[['cty', 'hwy']].std()
df_z
```

## better way

```python
def z (df):
    return (df - df.mean())/df.std()
```

```python
dz = z(df[['cty', 'hwy']])
dz
```

```python
pd.concat([df, dz], axis=1)
```

## use 'merge' or 'join' to join two dataframe and add suffix to overlapping columns names 

```python
df.join(dz, rsuffix='_z')
```

```python
df = df.merge(dz, left_index=True, right_index=True, suffixes=('','_z'))
```

```python
import pandas as pd
```

```python
df = pd.read_csv('https://github.com/prasertcbs/tutorial/raw/master/mpg.csv')
df
```

```python
df['rank'] = df.cty.rank(ascending=False)
```

```python
df
```

```python
df.sort_values('rank')
```

```python
df['rank'] = df.cty.rank(ascending=False, method='dense').astype(int)
df.sort_values('rank')
```

```python
df['rank'] = df.groupby('class').cty.rank(ascending=False, method='dense').astype(int)
df.sort_values(['class', 'rank'])
```

```python
df['tran_am'] = df.trans.str[0]
```

```python
df
```

```python
df['rank'] = df.groupby(['class', 'tran_am']).cty.rank(ascending=False, method='dense').astype(int)
df.sort_values(['class', 'tran_am', 'rank'])
```

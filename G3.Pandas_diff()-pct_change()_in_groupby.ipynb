```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('https://github.com/prasertcbs/tutorial/raw/master/eod1.csv')
df
```

```python
df.Close.diff()
```

```python
df['chg'] = df.groupby('Symbol').Close.diff()
```

```python
df
```

```python
df['pct_chg'] = df.groupby('Symbol').Close.pct_change()
```

```python
df
```

log return = $log(\frac{p_t}{p_{t-1}}) = ln(p_1) - ln(p_{t-1})$

```python
df['ln_return'] = df.groupby('Symbol').Close.apply(lambda v : np.log(v).diff())
```

```python
import pandas as pd
import numpy as np
%matplotlib inline
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/prasertcbs/tutorial/master/mpg.csv')
df
```

```python
df.info()
```

```python
df.pivot_table(index='class', columns='drv', values='model', aggfunc='count')
```

```python
df.pivot_table(index='class', columns='drv', values='cty', aggfunc='count', fill_value=0,    #fill_value เติมตัว Nan = 0
               margins=True, margins_name='Total').astype('int') 
```

```python
df.pivot_table(index='class', columns='drv', values='cty', aggfunc='count',
               fill_value=0).astype(int).plot.barh();
```

```python
df.pivot_table(index='class', columns='drv', values=['cty', 'hwy'], fill_value=0).round(2)
```

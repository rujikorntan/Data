## function

```python
def split2(df, train_ratio=0.7):
    ''''''
    df_train = df.sample(frac=train_ratio)
    df_test  = df[~df.index.isin(df_train.index)]
    return [df_train, df_test]
```

```python
df_train, df_test = split2(df)
```

```python
df_train
```

```python
df_test
```

## split 3 dataframe

```python
df_train2, df_temp = split2(df, train_ratio=.6)
df_validation2, df_test2 = split2(df_temp, train_ratio=0.5)
```

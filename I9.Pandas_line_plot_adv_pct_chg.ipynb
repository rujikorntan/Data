```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter
%matplotlib inline
%config InlineBlackend.figure_format = 'retina'
```

```python
pd.set_option('max_rows',10)
```

```python
df = pd.read_csv('https://github.com/prasertcbs/basic-dataset/raw/master/oil_gold_setindex.csv',
                 index_col='Date', parse_dates=True)
df
```

## % change form previous period

```python
df['oil_pct_chg'] = df.Oil.pct_change()
df['gold_pct_chg'] = df.Gold.pct_change()
df['set_pct_chg'] = df.SET.pct_change()
```

```python
df
```

```python
cols = ['oil_pct_chg', 'gold_pct_chg', 'set_pct_chg']
color = ['maroon', 'gold', 'deepskyblue', ]
```

```python
df[cols].plot()
```

```python
ax = df[cols].plot()
ax.axhline(0, linestyle='--', color='red', linewidth=2)
```

```python
ax = df.loc[::12, cols].plot()
ax.yaxis.set_major_formatter(FuncFormatter(lambda v, p:"{:,.0f}%".format(v*100)))     # v=value, p=position
```

```python
df[cols].plot(subplots=True, color=color, figsize=(6, 8))
```

```python
df[cols].plot(color=color, subplots=True, layout=(1, 3), figsize=(12,4))
```

## คำนวณ % การเปลี่ยนแปลงเทียบกับวันที่ที่กำหนดเป็นฐาน

```python
base_date = '2003-01-31'
```

```python
df.Oil.at[base_date]
```

```python
df['oil_pct_diff'] =df['Oil'] / df['Oil'].at[base_date] - 1
df['gold_pct_diff'] =df['Gold'] / df['Gold'].at[base_date] - 1
df['set_pct_diff'] =df['SET'] / df['SET'].at[base_date] - 1
```

```python
df
```

```python
ax = df.loc[base_date:, ['oil_pct_diff', 'gold_pct_diff', 'set_pct_diff']].plot(color=color)
ax.yaxis.set_major_formatter(FuncFormatter(lambda v, p: "{:,.0f}%".format(v*100)));
```

```python
import pandas as pd
```

```python
df = pd.read_csv('https://github.com/prasertcbs/tutorial/raw/master/mpg.csv')
df
```

```python
df.describe()
```

```python
df.groupby('class').describe()
```

```python
df.groupby('class').groups
```

```python
df.groupby('class').get_group('minivan')
```

```python
for k, v in df.groupby('class'):
    print(k)
    print('-----\n',v)
```

```python
df.groupby('class').mean()
```

```python
df.groupby('class').mean()[['displ']]
```

```python
%matplotlib inline
df.groupby('class').mean()[['displ','cty']].plot.barh()
```

```python
df.groupby('class').mean()[['cty', 'hwy']].sort_values('hwy').plot.barh()
```

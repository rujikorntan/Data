```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('https://github.com/prasertcbs/tutorial/raw/master/mpg.csv')
df.sample(10)
```

```python
df.plot.scatter(x='displ',y='cty');
```

```python
%config InlineBackend.figure_format = 'retina'    #ความละเอียดกราฟ
```

```python
df.plot.scatter(x='displ', y='cty')
```

```python
df.plot.scatter(x='displ', y='cty', title= 'City miles per gallon vs. engine size');
```

```python
df.plot.scatter(x='displ', y='cty', 
                c = df.drv.map({'4':'green', 'f':'blue', 'r':'red'}),             #color
                title= 'City miles per gallon vs. engine size');
```

```python
df.plot.scatter(x='displ', y='cty', 
                grid = True,
                c = df.drv.map({'4':'green', 'f':'blue', 'r':'red'}),  #color
                s = df.cyl*10,     #ขนาดจุดเพิ่มตาม กำลังสูบ
                title= 'City miles per gallon vs. engine size');
```

```python
df.plot.scatter(x='displ', y='cty', 
                grid = True,
                c = df.drv.map({'4':'green', 'f':'blue', 'r':'red'}),  #color
                s = df.cyl*10,     #ขนาดจุด
                alpha = 0.5,
                ylim=(0,40),
                xlim=(0,8),
                title= 'City miles per gallon vs. engine size');
```

```python
dfs = pd.read_html('https://en.wikipedia.org/wiki/Government_debt', match='country', header=0) #ชื่อ column
df = dfs[0]
df
```

```python
df.info()
```

```python
df['% of GDP'] = df['% of GDP'].str.replace(r'%', '').astype('float')
df
```

```python
df.info()
```

```python
df['% of world public debt'] =df['% of world public debt'].str.replace(r'%', '').astype('float')
df
```

```python
df.info()
```

```python
df['country'] = df['country'].str.replace(r'[^\s+ \w]','')        # ^ = ที่ไม่ใข่ \w = ตัวอักษร  \s =white space    
df
```

## ---------- Plot graph --------------------------

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
%config InlineBackend.figure_format = 'retina'
```

```python
dg = df.sort_values(by='% of GDP', ascending=False)
dg.head()
```

```python
plt.figure(figsize=(7, 7));
sns.barplot(x='% of GDP', y='country',
           data=dg, 
           color='.75');
```

```python
colors = ['deepskyblue' if r=='Thailand' else '0.75' for r in dg['country']]
colors
```

```python
plt.figure(figsize=(7, 7));
sns.barplot(x='% of GDP', y='country',
           data=dg, 
            palette = colors);
```

```python
colors = ['deepskyblue'if r=='Thailand' else '0.75' if r!='World' else 'green' for r in dg['country']]
```

```python
plt.figure(figsize=(7, 7))
sns.barplot(x='% of GDP', y='country',
           data=dg,
           palette = colors)
```

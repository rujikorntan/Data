```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format = 'retina'  #ความละเอียด 2เท่า
```

```python
df = pd.read_csv('https://github.com/prasertcbs/tutorial/raw/master/mpg.csv')
df.sample(6)
```

```python
df.drv.value_counts().plot(kind='barh');
```

```python
plt.style.use('ggplot')
df.drv.value_counts().plot.barh();
```

```python
plt.style.available
```

```python
plt.style.use('seaborn-pastel')
df.drv.value_counts().plot.barh();
```

```python
for s in sorted(plt.style.available):
    plt.style.use('default')
    plt.style.use(s)
    df.drv.value_counts().plot.barh(title=s, figsize=(4,2))
    plt.show()
```

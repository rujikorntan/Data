```python
import pandas as pd
```

```python
df = pd.read_csv('https://github.com/prasertcbs/tutorial/raw/master/mpg.csv')
df
```

```python
df = df.sample(10).reset_index(drop=True)
```

```python
df
```

```python
df.manufacturer = df.manufacturer.str.capitalize()
```

```python
df
```

```python
df['trans_am']=df.trans.map(lambda t : 'Auto' if t[0]=='a'  else 'Manual')
```

```python
df
```

```python
def am(v):
    if v[0] =='a':
        return 'Auto'
    else :
        return 'Manual'
```

```python
df.trans.map(lambda v: am(v))    #df.trans.map(am)
```

```python
df['tran_drv']=df.drv.map({'f':'Front', 'r':'Rear', '4':'4-Wheel'}).astype('category')
```
